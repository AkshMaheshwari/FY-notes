```python
password = input("Enter a password : ")

class InvalidLengthError(Exception):
 pass
class SpecialCharError(Exception):
 pass
class CaseExceptionError(Exception):
 pass
try:
 if len(password)<9:
  raise InvalidLengthError
 elif "!" not in password and "@" not in password and "#" not in password and "$" not in password and "%" not in password and "&" not in password and "*" not in password:
  raise SpecialCharError
 upper = False
 lower = False
 for i in range(0,len(password)):
  if password[i].isupper():
   upper = True
  if password[i].islower():
   lower = True

 if lower == False or upper == False:
  raise CaseExceptionError

 print("New password is valid: ",password)

except InvalidLengthError:
 print("Error : Length is too short")
except SpecialCharError:
 print("Error : No special characters in string")
except CaseExceptionError:
 print("Error : No upper lower case character combinations")
```
