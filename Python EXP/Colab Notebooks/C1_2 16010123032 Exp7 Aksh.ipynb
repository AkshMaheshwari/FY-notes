```python
#Q3)
import numpy as np
monthly_high_temperatures = np.array([25, 28, 30, 24, 23, 21, 32, 22, 28, 29, 31, 33])
monthly_low_temperatures = np.array([50, 45, 40, 38, 37, 41, 39, 36, 46, 40, 48, 55])
monthly_average_temperatures = (monthly_high_temperatures + monthly_low_temperatures) / 2
overall_average_high = np.mean(monthly_high_temperatures)
overall_average_low = np.mean(monthly_low_temperatures)
highest_avg_temp_month = np.argmax(monthly_average_temperatures) + 1
lowest_avg_temp_month = np.argmin(monthly_average_temperatures) + 1
print("Monthly High Temperatures:", monthly_high_temperatures)
print("Monthly Low Temperatures:", monthly_low_temperatures)
print("Monthly Average Temperatures:", monthly_average_temperatures)
print("Overall Average High Temperature:", overall_average_high)
print("Overall Average Low Temperature:", overall_average_low)
print("Month with the Highest Average Temperature:", highest_avg_temp_month)
print("Month with the Lowest Average Temperature:", lowest_avg_temp_month)



```
