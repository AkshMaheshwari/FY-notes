```python
letter_count = {}

# Get input from the user
input_string = input("Enter a string: ")

# Convert the input string to lowercase (optional, to treat 'A' and 'a' as the same letter)
input_string = input_string.lower()

# Iterate through each character in the input string
for char in input_string:
    # Check if the character is a letter (ignore spaces, numbers, and symbols)
    if char.isalpha():
        # If the character is already in the dictionary, increment its count by 1
        if char in letter_count:
            letter_count[char] += 1
        # If the character is not in the dictionary, add it with a count of 1
        else:
            letter_count[char] = 1

# Print the letter frequencies
for letter, count in letter_count.items():
    print(f"{letter}: {count}")
```
