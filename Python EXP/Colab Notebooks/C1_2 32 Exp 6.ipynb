```python
def get_employee_details():
    emp_id = input("Enter Employee ID: ")
    emp_name = input("Enter Employee Name: ")
    emp_department = input("Enter Employee Department: ")
    return emp_id, emp_name, emp_department

def write_to_file(emp_id, emp_name, emp_department, filename="employeedetails.txt"):
    with open(filename, "a") as file:
        file.write(f"{emp_id} {emp_name} {emp_department}\n")

while True:
    emp_id, emp_name, emp_department = get_employee_details()
    write_to_file(emp_id, emp_name, emp_department)

    another = input("Do you want to add another employee (yes/no)? ").strip().lower()
    if another =="yes":
      continue
    else:
      break


print("Employee details have been added to employeedetails.txt.")
```
